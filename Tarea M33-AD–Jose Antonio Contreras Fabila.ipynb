### 007 Tratamiento de Nulos

```python
# Se observan nulos en año puesto que el count es diferente al de las demás columnas
# Por lo que se identifican las filas con nulos
nulos_year = df[df['Year'].isnull()]
print(nulos_year)
```

```python
# Se da tratamiento de nulos NaN en la columna Year colocando 2006 la media como referencia
df['Year'] = df['Year'].fillna(2006).astype('float64')
```

```python
# Nuevamente se verifica la descripción de los datos y en count aparecen 16,598 registros en todas las columnas
df.describe()
```

```python
# Verificamos si existen nulos en alguna otra columna, encontramos que en Publisher si hay Nulos
print(df.isnull().any())
```

```python
# Número de datos vacíos por columna
missing_values_count = df.isnull().sum()
missing_values_count[0:10]
```

```python
df.info()
```

```python
# Conversión de las columnas a datos útiles
df['Name'] = df['Name'].astype('string')
df['Platform'] = df['Platform'].astype('string')
df['Year'] = df['Year'].astype('int64')
df['Genre'] = df['Genre'].astype('string')
df['Publisher'] = df['Publisher'].astype('string')
```

## Ventas

### Total de Ventas por Año

```python
# Se muestra el total de ventas por año
df.groupby('Year')['Global_Sales'].sum().reset_index().sort_values(by='Global_Sales', ascending=False)
```

### Juegos Más Vendidos por Año

```python
# Juegos más vendidos por año
juegos_top_por_año = (
    df.loc[df.groupby('Year')['Global_Sales'].idxmax(), ['Year', 'Name', 'Global_Sales']]
    .sort_values(by='Global_Sales', ascending=False)
    .reset_index(drop=True)
)
print(juegos_top_por_año)
```

### Plataformas con más ingresos por año

```python
# Agrupamos por año y plataforma, sumar ventas
ventas_por_plataforma = (
    df.groupby(['Year', 'Platform'])['Global_Sales']
      .sum()
      .reset_index()
)
# Obtener la plataforma con más ingresos por año
top_plataforma_por_año = (
    ventas_por_plataforma.loc[
        ventas_por_plataforma.groupby('Year')['Global_Sales'].idxmax(),
        ['Year', 'Platform', 'Global_Sales']
    ]
    .sort_values(by='Global_Sales', ascending=False)
    .reset_index(drop=True)
)

print(top_plataforma_por_año)
```

### Gráfico de comparación de ventas entre géneros por año

```python
# Agrupar por año y género, sumar ventas
ventas_por_genero = (
    df.groupby(['Year', 'Genre'])['Global_Sales']
      .sum()
      .reset_index()
)
# Gráfico de comparación
plt.figure(figsize=(12,6))
sns.lineplot(data=ventas_por_genero, x='Year', y='Global_Sales', hue='Genre', marker="o")
plt.title("Comparación de Ventas Globales por Género y Año")
plt.xlabel("Año")
plt.ylabel("Ventas Globales (millones)")
plt.legend(title="Género", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### InSights: Dos gráficos adicionales interesantes

```python
# Generar gráfico para ver qué plataformas lograron más ventas
# Agrupar por plataforma, sumar ventas Global_Sales
ventas_por_plataforma = (
    df.groupby('Platform')['Global_Sales']
      .sum()
      .reset_index().sort_values(by='Global_Sales', ascending=False)
)
# Crear gráfico comparativo
plt.figure(figsize=(12,6))
sns.barplot(data=ventas_por_plataforma, x='Platform', y='Global_Sales', hue='Platform')
plt.title("Ventas Globales por Plataforma")
plt.xlabel("Plataforma")
plt.ylabel("Ventas Globales (millones)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráfico comparativo de los 15 juegos más vendidos por año
juegos_top15_por_año = (
    df.loc[df.groupby('Year')['Global_Sales'].idxmax(), ['Year', 'Name', 'Global_Sales']]
    .sort_values(by='Global_Sales', ascending=False)
    .reset_index(drop=True)[0:15]
)
plt.figure(figsize=(14,7))
sns.barplot(data=juegos_top15_por_año, x='Year', y='Global_Sales', hue='Name')
plt.title("Top 15 Juegos Más Vendidos por Año (Global Sales)")
plt.xlabel("Año")
plt.ylabel("Ventas Globales (millones)")
plt.legend(title="Juego", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
# Gráfico de correlación
plt.Figure(figsize=(12,8))
corr_df = df.select_dtypes(include=['number']).corr(method='pearson')
df_lt = corr_df.where(np.tril(np.ones(corr_df.shape)).astype(bool))
hmap = sns.heatmap(df_lt, cmap='YlGnBu', annot=True)
```

### qué género ha tenido el mayor y menor crecimiento (últimos 4 años)

```python
# Nuevo DataFrame de ventas por gémero por año, últimos 4 años
df_filtrado = df[df['Year'] > 2016]
dfgenre = pd.pivot_table(df_filtrado, values='Global_Sales', index=['Genre'], columns=['Year'], aggfunc=sum).fillna(0).reset_index()
dfgenre # El resultado sugiere que Simulation es el de mayor crecimiento y el de menor crecimiento ha sido Action
```

```python
# Gráfico de barras
dfgenre.set_index('Genre').plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    colormap='tab20'
)

plt.title("Ventas Globales por Género y Año (>2016)")
plt.xlabel("Género")
plt.ylabel("Ventas Globales")
plt.legend(title="Año")
plt.tight_layout()
plt.show()
```

### 2 elementos adicionales que muestren oportunidades de mercado

```python
# Se genera un DataFrame con las ventas por región y género
ventas_region_genero = (
    df.groupby(["Genre"])[["NA_Sales","EU_Sales","JP_Sales","Other_Sales"]]
      .sum()
      .reset_index()
)
print("Ventas por región y género:\n", ventas_region_genero.head(15))
```

```python
# Configuración del gráfico
plt.figure(figsize=(12,6))
ventas_melt = ventas_region_genero.melt(id_vars="Genre",
                                        value_vars=["NA_Sales","EU_Sales","JP_Sales","Other_Sales"],
                                        var_name="Región",
                                        value_name="Ventas")

# Gráfico de barras agrupadas
sns.barplot(data=ventas_melt, x="Genre", y="Ventas", hue="Región")
plt.title("Ventas globales por género y región")
plt.xticks(rotation=45)
plt.ylabel("Ventas (millones)")
plt.xlabel("Género")
plt.legend(title="Región")
plt.tight_layout()
plt.show()
```
